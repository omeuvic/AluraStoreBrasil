### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

todas_lojas = pd.concat([loja, loja2, loja3, loja4])

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja['Preço'].sum()
print("O faturamento da loja 1 foi de",faturamento_loja1)
faturamento_loja2 = loja2['Preço'].sum()
print("O faturamento da loja 2 foi de",faturamento_loja2)
faturamento_loja3 = loja3['Preço'].sum()
print("O faturamento da loja 3 foi de",faturamento_loja3)
faturamento_loja4 = loja4['Preço'].sum()
print("O faturamento da loja 4 foi de",faturamento_loja4)

faturamento_total = sum([
    faturamento_loja1,
    faturamento_loja2,
    faturamento_loja3,
    faturamento_loja4
])
print("O faturamento total foi de",faturamento_total)

```

# 2. Vendas por Categoria

```python



vendas_por_categoria = loja.groupby('Categoria do Produto')['Preço'].sum()
print(vendas_por_categoria)

vendas_por_categoria2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
print(vendas_por_categoria2)

vendas_por_categoria3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
print(vendas_por_categoria3)

vendas_por_categoria4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
print(vendas_por_categoria4)
```

# 3. Média de Avaliação das Lojas


```python
soma_Avaliação = loja['Avaliação da compra'].sum()
media_avaliação = soma_Avaliação / len(loja['Avaliação da compra'])
print(f" loja 1 Média: {media_avaliação:.2f}")

soma_Avaliação = loja2['Avaliação da compra'].sum()
media_avaliação = soma_Avaliação / len(loja2['Avaliação da compra'])
print(f" loja 2 Média: {media_avaliação:.2f}")

soma_Avaliação = loja3['Avaliação da compra'].sum()
media_avaliação = soma_Avaliação / len(loja3['Avaliação da compra'])
print(f" loja 3 Média: {media_avaliação:.2f}")

soma_Avaliação = loja4['Avaliação da compra'].sum()
media_avaliação = soma_Avaliação / len(loja4['Avaliação da compra'])
print(f" loja 4 Média: {media_avaliação:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas = loja.groupby('Categoria do Produto').size()
vendas2 = loja2.groupby('Categoria do Produto').size()
vendas3 = loja3.groupby('Categoria do Produto').size()
vendas4 = loja4.groupby('Categoria do Produto').size()


print("loja 1",vendas)
print("loja 2",vendas2)
print("loja 3",vendas3)
print("loja 4",vendas4)

```

# 5. Frete Médio por Loja

```python
frete_medio = loja['Frete'].mean()

print(f"Frete médio loja 1: R$ {frete_medio:.2f}")

frete_medio2 = loja2['Frete'].mean()

print(f"Frete médio loja 2: R$ {frete_medio2:.2f}")

frete_medio3 = loja3['Frete'].mean()

print(f"Frete médio loja 3: R$ {frete_medio3:.2f}")

frete_medio4 = loja['Frete'].mean()

print(f"Frete médio loja 4: R$ {frete_medio4:.2f}")
```

6. GRAFICOS

BARRA LOJA 1

```python
avaliacao_categoria = loja.groupby('Categoria do Produto')['Avaliação da compra'].mean()

plt.figure()
plt.bar(avaliacao_categoria.index, avaliacao_categoria.values)
plt.xticks(rotation=45)
plt.title("Média de Avaliação por Categoria")
plt.ylabel("Média da Avaliação")
plt.show()
```

PIZZA LOJA 1

```python
quantidade_categoria = loja.groupby('Categoria do Produto').size()

plt.figure()
plt.pie(quantidade_categoria.values, labels=quantidade_categoria.index, autopct='%1.1f%%')
plt.title("Distribuição de Vendas por Categoria")
plt.show()
```

LINHA LOJA 1

```python
import matplotlib.pyplot as plt

vendas_categoria = loja.groupby('Categoria do Produto')['Preço'].sum()

plt.figure()
plt.plot(vendas_categoria.index, vendas_categoria.values, marker='o')
plt.xticks(rotation=45)
plt.title("Faturamento por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Total (R$)")
plt.show()
```
